### Function to split and add the values seperated by '+' operator.

```python
import numpy as np

def converter(x):
  if isinstance(x, str):
    token = x.split('+')
    if len(token)==2:
      return((float(token[0])+ float(token[1])))

  try:
    return float(x)
  except:
    return np.nan

```

```python
for i, c in enumerate(df2.columns):
  df2[c] = df2[c].apply(converter)
```

```python
for i, c in enumerate(df2.columns):
  df2.loc[df2[c].isnull(), c] = 0
```

###Function to handle Outliers

```python
plt.figure(figsize=(20,25))
plot=1
for i in df3.columns:
    plt.subplot(25,3,plot)
    plt.boxplot(x=df3[i])
    plt.xlabel(i,fontsize=15)
    plot+=1
plt.show()
```

```python
def outliers(df,column):
    Q1 = np.percentile(df[column], 25)
    Q3 = np.percentile(df[column], 75)
    IQR = Q3 - Q1
    LB = Q1 - (1.5 * IQR)
    UB = Q3 + (1.5 * IQR)

    mean_value = df[column].mean()

    df[column] = np.where((df[column] < LB) | (df[column] > UB), mean_value, df[column])

    return df

for i in df3.columns:
    df3=outliers(df3,i)
```

### Scaling the data using Satndard Scalar method

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
df_skills_scale=scaler.fit_transform(df_skills)
```

## **Applying PCA Method**

```python
from sklearn.decomposition import PCA
pca=PCA()
p_c=pca.fit_transform(df_skills_scale)
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_),marker="o")
plt.xlabel('Number of Components')
plt.ylabel('Variance (%)')
plt.title('Explained Variance')
plt.show()
```

* From the above graph we are retaining the 94% varience, which corresponds to 15 Principle Components.

```python
pca=PCA(n_components=15)
p_c=pca.fit_transform(df_skills_scale)
```

```python
pca.explained_variance_ratio_
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
p_c=pd.DataFrame(p_c)
p_c.columns=[f'PC{i+1}' for i in range(p_c.shape[1])]
```

```python
p_c.head()
```

Applying K-Means after PCA analysis

```python
from sklearn.cluster import KMeans
see=[]
k_rng=range(1,11)
for k in k_rng:
    km=KMeans(n_clusters=k,init='k-means++',random_state=42)
    km.fit(p_c)
    see.append(km.inertia_)
```

```python
plt.plot(k_rng,see)
plt.xlabel('k')
plt.ylabel('see')
```

* From the above graph the elbow clearly corresponds to 4 Clusters.

```python
km=KMeans(n_clusters=4,init='k-means++',random_state=42)
km.fit(p_c)
```

## **K-Means**

```python
df_skills_scale.shape
```

```python
from sklearn.cluster import KMeans
see=[]
k_rng=range(1,11)
for k in k_rng:
    km=KMeans(n_clusters=k,init='k-means++',random_state=42)
    km.fit(df_skills_scale)
    see.append(km.inertia_)
```

```python
plt.plot(k_rng,see)
plt.xlabel('k')
plt.ylabel('see')
```

* Elbow method to determine N value for clustering

```python
from sklearn.metrics import silhouette_score

silhouette_scores = []
for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(df_skills_scale)
    score = silhouette_score(df_skills_scale, kmeans.labels_)
    silhouette_scores.append(score)

plt.plot(range(2, 11), silhouette_scores)
plt.title('Silhouette Score Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.show()
```

* Silhouette Score Method to determine N value for the cluster. From the Visualization we can observe 4 is the highest value.
